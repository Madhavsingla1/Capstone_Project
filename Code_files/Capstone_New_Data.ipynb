```python
import pandas as pd
import numpy as np
import xml.etree.ElementTree as et
import codecs
import csv
import time
import os
```

```python
def strip_tag_name(t):
    t = elem.tag
    # print(t)
    idx = k = t.rfind("}")
    if idx != -1:
        t = t[idx + 1:]
    return t
```

```python
# ENCODING = "utf-8"
# with codecs.open("/content/drive/MyDrive/Capstone Data/NBestanswers.csv", "w", ENCODING) as articlesFH:
#   nbestanswer = csv.writer(articlesFH, quoting=csv.QUOTE_MINIMAL)
  # DataSet = csv.writer(Data,quoting=csv.QUOTE_MINIMAL)
cat=[]
maincat=[]
subject=[]
bestanswer=[]
# nbestanswer=[]
answer=[]
content=[]
id =[]
bestid = []
document=[]
mx=0
# totalCount=0;
for event, elem in et.iterparse("/content/drive/MyDrive/Capstone Data/FullOct2007.xml", events=('start', 'end')):
        tname = strip_tag_name(elem.tag)
        if event == 'start':
            if tname == 'cat':
              cat.append(elem.text)
            elif tname == 'maincat':
              maincat.append(elem.text)
            elif tname == 'subject':
              subject.append(elem.text)
            elif tname == 'bestanswer':
              bestanswer.append(elem.text)
            elif tname == 'answer_item':
              answer.append(elem.text)
            elif tname == 'content':
              content.append(elem.text)
            elif tname == 'id':
              id.append(elem.text)
            elif tname == 'best_id':
              bestid.append(elem.text)
            elif tname=='document':
              document.append(elem.get('type',''))
        elif tname=='vespaadd':
        #   nbestanswer.writerow(answer[:])
          mx=max(mx,len(answer))
          answer.clear()
        elem.clear()
```
